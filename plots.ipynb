```python
models = ["clear_transformer", "fullconnected_GATv2Conv", "fullconnected_GATConv"] #"fullconnected_GAT"
n_tests = 2
epoch = 10

data = {}

for model in models:
    tests = []
    for i in range(n_tests):
        times = []
        loss_list = []
        with open(f"results/{model}_{i}") as f:
            lines = f.readlines()
            for line in lines:
                line = line.split(";")
                times.append(float(line[1]))
                loss_list.append(float(line[0]))
        tests.append((times, loss_list))
    
    data[model] = tests
```

```python
print(data)
```

```python
import matplotlib.pyplot as plt
import numpy as np

for model in models:
    loss_list = np.zeros(epoch)
    for i in range(n_tests):
        loss_list+= np.array(data[model][i][1])
    loss_list/=n_tests
    
    plt.plot(loss_list)


```
